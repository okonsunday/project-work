## Cluster Analysis Assignment 

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
from sklearn.cluster import KMeans
```

```python
raw_data = pd.read_csv("3.12.+Example.csv")
data= raw_data.copy()
data
```

## Clustering based on Loyalty

```python
X = data.iloc[:, 0:2]
X
```

```python
vincent = KMeans(n_clusters= 4)
vincent.fit(X)
```

```python
ic = vincent.fit_predict(X)
ic
```

```python
x = data[ "Satisfaction"]
y = data["Loyalty"]
plt.scatter(x,y, c= ic, cmap="rainbow")
plt.xlim(0, 20)
plt.ylim(-0.001, 5 )
plt.title("SATISFACTION VS LOYALTY")
plt.xlabel("satisfaction", fontsize =15)
plt.ylabel("loyalty", fontsize = 15)
plt.show()
```
