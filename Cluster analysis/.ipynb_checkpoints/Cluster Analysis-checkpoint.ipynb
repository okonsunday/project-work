```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

import seaborn as sns
sns.set()

from sklearn.cluster import KMeans
```

```python
# Load data

raw_data = pd.read_csv("3.01.+Country+clusters.csv")
data = raw_data.copy()
data
```

### Clustering based on Geo Location

```python
X = data.iloc[:, 1:3]
X
```

```python
vincent = KMeans(n_clusters=4)
vincent.fit(X)
```

```python
# identified_clusters
ic = vincent.fit_predict(X)
ic
```

```python
# Whenever you are plotting geographical location, 
# Longitude represents x-axis
# Latitude represents y-axis

x = X.iloc[:, 1]
y = X.iloc[:, 0]


plt.scatter(x, y, c=ic, cmap="rainbow")
plt.xlim(-150, 180)
plt.ylim(-50, 100)

plt.title("K-Means Clustering")
plt.xlabel("Longitude", fontsize=15)
plt.ylabel("Latitude", fontsize=15)
plt.show()
```

## Elbow-Method WCSS

```python
# Within-cluster sum of squares


import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.cluster import KMeans
```

```python
raw_data = pd.read_csv("3.01.+Country+clusters.csv")

data = raw_data.copy()
data
```

```python
X = data.iloc[:, 1:3]
X
```

```python
# Whenever you are plotting geographical location, 
# Longitude represents x-axis
# Latitude represents y-axis

plt.scatter(data["Longitude"], data["Latitude"])
```

```python
kmeans = KMeans(n_clusters=2)
kmeans.fit(X)
```

```python
ic = kmeans.fit_predict(X)
ic
```

```python
# Whenever you are plotting geographical location, 
# Longitude represents x-axis
# Latitude represents y-axis

plt.scatter(x=data["Longitude"], y=data["Latitude"], c=ic)
```

```python
wcss = kmeans.inertia_
wcss
```

```python
# kmeans = KMeans(n_clusters=4)

wcss = []

for i in range(1, 9):
    kmeans = KMeans(n_clusters=i)
    kmeans.fit(X)
    wcss_iter = kmeans.inertia_
    wcss.append(wcss_iter)
    
```

```python
wcss
```

```python
# Plotting the elbow method

number_cluster = range(1, 9)

plt.plot(number_cluster, wcss)
```
